```python
def format_links(link):
    if not link.startswith(('https://www', "http://www.", "http://", "https://")):
        if not link.startswith(('www.')):
            link = 'www.' + link
        if not link.startswith(('http://', 'https://')):
            link = 'https://' + link
    return link
```

```python
def modify_input():
    with open('science.json') as file:
        a = json.load(file)
    with open('conspiracy.json') as file:
        b = json.load(file)
    rev_a = []
    rev_b = []
    for a_it in a:
        print(a_it)
        rev_a.append({'title': a_it[0], 'link': format_links(a_it[1])})
    for b_it in b:
        rev_b.append({'title': b_it[0], 'link': format_links(b_it[1])})
    with open('science.json', 'w') as file:
        json.dump(rev_a, file, indent=4)
    with open('conspiracy.json', "w") as file:
        json.dump(rev_b, file, indent=4)
```

```python
modify_input()
```

```python
# adds new articles discovered to existing database (extending the database)
def update_data(new_data, article_type):
    def remove_dup(curr):
        titles = set()
        rev = []
        for item in curr:
            if item['title'] in titles:
                continue
            rev.append(item)
            titles.add(item['title'])
        return rev

    if not (article_type == "science" or article_type == "conspiracy"):
        raise ValueError("article_type param must either be 'science' or 'conspiracy'")

    with open(f"data/extract_articles/{article_type}.json") as file:
        existing_data = json.loads(file.read())

    # print(len(existing_data))
    existing_data += new_data
    # print(len(existing_data))

    with open(f"data/extract_articles/{article_type}.json", 'w') as file:
        file.write(json.dumps(remove_dup(existing_data), indent = 4))
```

```python
with open("science.json") as file:
    old_science_data = json.loads(file.read())

with open("conspiracy.json") as file:
    old_conspiracy_data = json.loads(file.read())
```

```python
update_data(old_science_data, article_type="science")
```

```python
update_data(old_conspiracy_data, article_type="conspiracy")
```
